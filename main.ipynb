```python
data = pd.read_csv('titanic.csv')
fig = px.bar(data, x='Sex', y='Survived', color='Survived', text='Survived', barmode='group', 
             labels={'Survived': 'Number of Passengers'},
             title='Survival Rate by Gender')
fig.show()
```

```python

fig = px.bar(data, x='Sex', y='Survived', color='Survived', text='Survived', barmode='group', 
             labels={'Survived': 'Number of Passengers'},
             title='Survival Rate by Gender')
fig.show()
```

```python
fig = px.histogram(data, x='Age', color='Survived', nbins=20, 
                   labels={'Age': 'Age of Passengers', 'count': 'Number of Passengers'},
                   title='Age Distribution of Passengers by Survival Status')
fig.show()
```

```python
fig = px.violin(data, x='Pclass', y='Fare', box=True, points='all', 
                labels={'Pclass': 'Passenger Class', 'Fare': 'Fare Price'},
                title='Fare Price Distribution by Passenger Class')
fig.show()
```

```python
fig = px.scatter(data, x='Age', y='Fare', color='Pclass', symbol='Pclass',
                 labels={'Age': 'Age of Passengers', 'Fare': 'Fare Price', 'Pclass': 'Passenger Class'},
                 title='Age vs. Fare Price by Passenger Class')
fig.show()
```

```python
fig = px.scatter_matrix(data, dimensions=['Age', 'Fare', 'Pclass'], color='Survived',
                        labels={'Age': 'Age of Passengers', 'Fare': 'Fare Price', 'Pclass': 'Passenger Class', 'Survived': 'Survival Status'},
                        title='Scatter Plot Matrix of Age, Fare, and Class by Survival Status')
fig.show()
```

```python
pio.write_html(fig, './html_img/bar_plot.html')
pio.write_html(fig, './html_img/histogram_plot.html')
pio.write_html(fig, './html_img/violin_plot.html')
pio.write_html(fig, './html_img/scatter_plot.html')
pio.write_html(fig, './html_img/scatter_matrix.html')
```
